```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

# Amostragem estratificada

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
iris['class'].value_counts()
```

```python
X = iris.iloc[:, 0:-1]
Y = iris.iloc[:, -1]

X
```

```python
# stratify: para retornar a amostra baseada na classe
X_train, Y_train, X_test, Y_test = train_test_split(X, Y, test_size = 0.3, stratify = Y)
X_train
```

```python
infert = pd.read_csv('infert.csv')
infert
```

```python
infert['education'].value_counts()
```

```python
X = infert.iloc[:, 2:9]
Y = infert.iloc[:, 1]
```

```python
# Criando uma amostra com somente 40% dos registros (por isso é definido 0.6, pois é gerado o inverso)
X_1, _, Y_1, _ = train_test_split(X, Y, test_size = 0.6, stratify = Y)
Y_1.value_counts()
```
