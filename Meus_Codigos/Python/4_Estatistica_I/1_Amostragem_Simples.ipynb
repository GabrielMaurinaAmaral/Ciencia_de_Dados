```python
base = pd.read_csv('iris.csv')
base
```

```python
# 150 amostras, de 0 a 1, com reposição, probabilidades equivalentes
amostra = np.random.choice(a = [0, 1], size = 150, replace = True,
                           p = [0.7, 0.3])


print(len(amostra))
print(len(amostra[amostra == 1]))
print(len(amostra[amostra == 0]))
amostra
```
