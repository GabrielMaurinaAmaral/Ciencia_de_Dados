```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Divisão da tela (subgráficos)

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
plt.scatter(base.Girth, base.Volume)
```

```python
plt.scatter(base.Girth, base.Height)
```

```python
plt.scatter(base.Height, base.Volume, marker = '*')
```

```python
plt.hist(base.Volume)
```

```python
# Criação de figura, no qual os gráficos serão posicionados
plt.figure(1)

plt.subplot(2, 2, 1)
plt.scatter(base.Girth, base.Volume)
plt.title('Girth vs Volume')

plt.subplot(2, 2, 2)
plt.scatter(base.Girth, base.Height, marker=11)
plt.title('Girth vs Height')

plt.subplot(2, 2, 3)
plt.scatter(base.Height, base.Volume, marker='*')
plt.title('Height vs Volume')

plt.subplot(2, 2, 4)
plt.hist(base.Volume)
plt.title('Volume Histogram')

# Ajusta o layout para evitar sobreposição
plt.tight_layout()
plt.show()
```
