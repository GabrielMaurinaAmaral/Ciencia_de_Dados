# Histograma

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
np.histogram(base.iloc[:,1], bins = 8)
```

```python
# bins -> determina no munero de colunas que vai ter
plt.hist(base.iloc[:, 1], bins = 8)
plt.title('Árvores')
plt.ylabel('Frequência')
plt.xlabel('Altura')
```

```python
plt.hist(base.iloc[:, 2], bins = 8)
plt.title('Árvores')
plt.ylabel('Frequência')
plt.xlabel('Volume')
```

```python
plt.hist(base.iloc[:, 0], bins = 8)
plt.title('Árvores')
plt.ylabel('Frequência')
plt.xlabel('Girth')
```
