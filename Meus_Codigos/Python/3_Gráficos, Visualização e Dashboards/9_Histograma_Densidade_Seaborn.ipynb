```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

# Histograma com densidade e seabord

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Histograma com 10 divisões (bins) e com gráfico de densidade (kernel density - suaviação)
# stat: count, frequency, probability,percent, density
srn.histplot(base.Volume, color="blue", kde=True, stat="density", bins=8)
```

```python
base2 = pd.read_csv('chicken.csv')
base2.head()
```

```python
# Criação de novo dataframe agrupando o atributo 'feed'
agrupado = base2.groupby(['feed'])['weight'].sum()
agrupado
```

```python
#somente para testar os filtros do pandas
teste = base2.loc[base2['feed'] == 'horsebean']
teste
```

```python
# Histrograma considerando somente o valor 'horsebean'
srn.kdeplot(base2.loc[base2['feed'] == 'horsebean'].weight)
```

```python
# Histrograma considerando somente o valor 'casein'
srn.histplot(base2.loc[base2['feed'] == 'casein'].weight, kde=True, bins=8).set(title='casein')
```

```python
# Histrograma considerando somente o valor 'linseed'
srn.histplot(base2.loc[base2['feed'] == 'linseed'].weight, kde=True, bins=8).set(title='linseed')
```

```python
# Histrograma considerando somente o valor 'meatmeal'
srn.histplot(base2.loc[base2['feed'] == 'meatmeal'].weight, kde=True, bins=8).set(title='meatmeal')
```

```python
# Histrograma considerando somente o valor 'soybean'
srn.histplot(data=base2.loc[base2['feed'] == 'soybean'].weight, kde=True, bins=8).set(title='soybean')
```

```python
# Histrograma considerando somente o valor 'sunflower'
srn.histplot(base2.loc[base2['feed'] == 'sunflower'].weight, kde=True, bins=8).set(title='sunflower')
```

```python
#impresssão em gráfico 2 x 3 - 3 linhas 2 colunas
plt.figure()

plt.subplot(3,2,1)
srn.histplot(base2.loc[base2['feed'] == 'horsebean'].weight, kde=True).set(title='horsebean')

plt.subplot(3,2,2)
srn.histplot(base2.loc[base2['feed'] == 'casein'].weight, kde=True).set(title='casein')

plt.subplot(3,2,3)
srn.histplot(base2.loc[base2['feed'] == 'linseed'].weight, kde=True).set(title='linseed')

plt.subplot(3,2,4)
srn.histplot(base2.loc[base2['feed'] == 'meatmeal'].weight, kde=True).set(title='meatmeal')

plt.subplot(3,2,5)
srn.histplot(base2.loc[base2['feed'] == 'soybean'].weight, kde=True).set(title='soybean')

plt.subplot(3,2,6)
srn.histplot(base2.loc[base2['feed'] == 'sunflower'].weight, kde=True).set(title='sunflower')

plt.tight_layout()
```
