```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Gr√°ficos de densidade

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
plt.hist(base.iloc[:,1], bins = 8)
```

```python
#kde = linha de densidade
sns.histplot(base.iloc[:, 1], kde=False, bins=8, color='blue').set(title='√Årvores')
```

```python
#m√©todo kdeplot para densidade
sns.kdeplot(base.iloc[:, 1], color='blue').set(title='√Årvores')
```

```python
#densidade e histograma
sns.histplot(base.iloc[:, 1], kde=True, bins=8, color='blue').set(title='√Årvores')
```
