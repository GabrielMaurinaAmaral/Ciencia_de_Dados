# Boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# patch_artist = True preenche, showfliers outliers
plt.boxplot(base.Volume, vert = False, showfliers = False, notch = True, patch_artist = True)
plt.title('Árvores')
plt.xlabel('Volume')
```

```python
#dados por linha
plt.boxplot(base)
plt.title('Árvores')
plt.xlabel('Dados')
```

```python
# Criação dos subplots
fig, axs = plt.subplots(3)

# Plotagem dos boxplots
axs[0].boxplot(base.Volume, vert = False, notch = True, patch_artist = True)
axs[0].set_title('Volume')
axs[1].boxplot(base.Girth, vert = False, notch = True, patch_artist = True)
axs[1].set_title('Girth')
axs[2].boxplot(base.Height, vert = False, notch = True, patch_artist = True)
axs[2].set_title('Height')

# Ajuste do layout
plt.tight_layout()
plt.show()
```
