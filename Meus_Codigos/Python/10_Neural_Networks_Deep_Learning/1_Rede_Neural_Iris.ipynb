```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
# from keras.utils import np_utils
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img

imagem = img.imread('rede_neural.png')
plt.imshow(imagem)
plt.axis('off') 
plt.show()
```

# Redes neurais artificiais com keras

```python
base = datasets.load_iris()
previsores = base.data
classe = base.target
print(previsores, classe)
```

# TRANFORMANDO CLASSE COM MAIS 3 OPES 

```python
# temos uma rede neural com 3 neurnios na camada de sada
classe_dummy = np.eye(3)[classe]
classe_dummy
```

# SEPARANDO DADOS PARA TRATAMENTO

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores,
                                                                  classe_dummy,
                                                                  test_size = 0.3,
                                                                  random_state = 0)
```

# CRIANDO MODELO - REDE NEURAL

```python
# estrutura da rede neural com a classe Sequential (sequncia de camadas)
modelo = Sequential()
# primeira camada oculta, 5 neuronios, 4 neuronios de entrada
modelo.add(Dense(units = 5, input_dim = 4))
# segunda camada oculta
modelo.add(Dense(units = 4))
# Funo softmax porque temos um problema de classificao com mais de duas classes 
modelo.add(Dense(units = 3, activation = 'softmax'))

modelo.summary()
```

```python
# adam = algoritmo para atualizar os pesos --- loss = clculo do erro
modelo.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
# treinamento -> dividindo a base de treinamento para validao (validation_data)
modelo.fit(X_treinamento, y_treinamento, epochs = 1000, validation_data = (X_teste, y_teste))
```

# ANALISANDO RESULTADOS

```python
# True ou False de acordo com o threshold 0.5
previsoes = modelo.predict(X_teste)
previsoes = (previsoes > 0.5)
previsoes
```

```python
# Como  um problema com trs sadas, precisamos buscar a posio que possui o maior valor
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
y_teste_matrix
```

```python
# Gerao da matriz de confuso
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
confusao
```
