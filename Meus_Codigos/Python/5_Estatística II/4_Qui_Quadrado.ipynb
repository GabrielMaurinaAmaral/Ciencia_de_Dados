```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('qui_quadrado.jpg')
plt.imshow(img)
plt.axis('off') 
plt.show()
```

```python
# Criao da matriz com os dados e execuo do teste
novela = np.array([[19, 6], [43, 32]])
novela
```

```python
#segundo valor  o pvalue
#Valor de p  maior que 0,05 no h evidncias de diferena significativa (hiptese nula): no h diferena significativa
chi2_contingency(novela)
```

```python
novela2 = np.array([[22, 3], [43, 32]])
novela2
```

```python
#agora valor de p menor que 0,05, podemos rejeitar a hiptese nula em favor da hiptese alternativa: h diferena significativa
chi2_contingency(novela2)
```
