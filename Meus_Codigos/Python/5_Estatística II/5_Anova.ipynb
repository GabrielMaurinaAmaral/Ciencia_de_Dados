```python
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import MultiComparison
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('anova.png')
plt.imshow(img)
plt.axis('off') 
plt.show()
```

```python
tratamento = pd.read_csv('anova.csv', sep = ';')
tratamento.head()
```

```python
# Boxplot agrupando os dados pelo remdio
tratamento.boxplot(by = 'Remedio', grid = False)
```

```python
# Criao do modelo de regresso linear e execuo do teste
modelo1 = ols('Horas ~ Remedio', data = tratamento).fit()
resultados1 = sm.stats.anova_lm(modelo1)
# Observar valor de p maior que 0,05 (Pr(>F)) Hiptese nula de que no h diferena significativa
resultados1
```

```python
# Criao do segundo modelo utilizando mais atributos e execuo do teste
modelo2 = ols('Horas ~ Remedio * Sexo', data = tratamento).fit()
resultados2 = sm.stats.anova_lm(modelo2)
#Nenhum valor de P mostra diferena significativa
resultados2
```

```python
#Se houver diferena o teste de Tukey  executado
# Execuo do teste de Tukey e visualizao dos grficos com os resultados
mc = MultiComparison(tratamento['Horas'], tratamento['Remedio'])
resultado_teste = mc.tukeyhsd()
print(resultado_teste)
resultado_teste.plot_simultaneous()
```
